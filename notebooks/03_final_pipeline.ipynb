```python
import pandas as pd
from notebooks.01_regression_price_prediction import RegressionTrainer
from notebooks.02_classification_recommendation import ClassificationTrainer
```

```python

df = pd.read_csv("data/Used_car_price_Prediction.csv")
reg_trainer = RegressionTrainer(df, target="selling_price")
lr_model, lr_scores = reg_trainer.train_linear_regression()
rf_model, rf_scores = reg_trainer.train_random_forest()
gb_model, gb_scores = reg_trainer.train_gradient_boosting()

print("REGRESSION RESULTS")
print("Linear Regression:", lr_scores)
print("Random Forest:", rf_scores)
print("Gradient Boosting:", gb_scores)

df['price_category'] = pd.qcut(df['selling_price'], q=3, labels=[0, 1, 2])
df_class = df.drop("selling_price", axis=1)
cls_trainer = ClassificationTrainer(df_class, target="price_category")
log_model, log_scores = cls_trainer.train_logistic_regression()
clf_rf_model, clf_rf_scores = cls_trainer.train_random_forest()
clf_gb_model, clf_gb_scores = cls_trainer.train_gradient_boosting()

print("\nCLASSIFICATION RESULTS")
print("Logistic Regression:", log_scores)
print("Random Forest:", clf_rf_scores)
print("Gradient Boosting:", clf_gb_scores)
```
